```python
import numpy as np
import math
import os
import librosa
import matplotlib.pyplot as plt
import librosa.display
```

```python
classical_filepath = r"/content/drive/MyDrive/Thesis/Datasets/GTZAN Dataset/genres_original/classical/classical.00000.wav"
metal_filepath = r"/content/drive/MyDrive/Thesis/Datasets/GTZAN Dataset/genres_original/metal/metal.00000.wav"
disco_filepath = r"/content/drive/MyDrive/Thesis/Datasets/GTZAN Dataset/genres_original/disco/disco.00000.wav"
reggae_filepath = r"/content/drive/MyDrive/Thesis/Datasets/GTZAN Dataset/genres_original/reggae/reggae.00000.wav"
```

```python
cl_data, cl_sr = librosa.load(classical_filepath)
cl_mfcc = librosa.feature.mfcc(cl_data, sr = cl_sr, n_mfcc = 13)
plt.figure(figsize=(25, 10))
librosa.display.specshow(cl_mfcc, 
                         sr=cl_sr, 
                         x_axis="time")
plt.colorbar(format="%+2f")
```

```python
metal_data, metal_sr = librosa.load(metal_filepath)
metal_mfcc = librosa.feature.mfcc(metal_data, sr = metal_sr, n_mfcc = 13)
plt.figure(figsize=(25, 10))
librosa.display.specshow(metal_mfcc, 
                         sr=metal_sr, 
                         x_axis="time")
plt.colorbar(format="%+2f")
```

```python
disco_data, disco_sr = librosa.load(disco_filepath)
disco_mfcc = librosa.feature.mfcc(disco_data, sr = disco_sr, n_mfcc = 13)
plt.figure(figsize=(25, 10))
librosa.display.specshow(disco_mfcc, 
                         sr=disco_sr, 
                         x_axis="time")
plt.colorbar(format="%+2f")
```
